```python
# Q3. select * from cust where Income = 23

df[df['Income'] == 23]
```

```python
# Q4. select ID from cust where income=23

df[df['Income'] == 23]['ID']
```

```python
# Q5. select ID,Gender from cust where income=23

df[df['Income'] == 23][['ID', 'Gender']]
```

```python
# Q6. select * from cust where income=23 and Score=98

df[(df['Income'] == 23) & (df['Score'] == 98)]
```

```python
# Q7. select ID, Age from cust where Income=23 and Score=98

df[(df['Income'] == 23) & (df['Score'] == 98)][['ID', 'Age']]
```

```python
# Q8. select mean(Age),max(Age),min(Age) from cust

print("mean age : ", df['Age'].mean())
print("max age : ", df['Age'].max())
print("min age : ", df['Age'].mean())
```

```python
# Q9. select distinct Income from cust

np.unique(df['Income'])
```

```python
df.drop_duplicates('Income')['Income']
```

```python
# Q10. select * from cust where gender='Female' order by Score

df[df['Gender'] == 'Female'].sort_values(by = ['Score'])
```

```python
# Q11. select * from cust where gender='Female' order by Score desc

df[df['Gender'] == 'Female'].sort_values(by = ['Score'], ascending = False)
```

```python
# Q12. select Gender, Age, count(*) from cust group by Gender, Age

df.groupby(['Gender', 'Age']).agg('count')
```

```python
# Q13. select Gender, Age, count(*) from cust group by Gender, Age order by Age desc

df.groupby(['Gender', 'Age']).agg('count').sort_values(by='Age')
```

```python
# Q14. select * from cust where Age in(20,30,40)

df[(df['Age'] >= 20) & (df['Age'] <= 40 )]
```

```python
# Q15. select * from cust where Age not in(20,30,40)

df[(df['Age'] < 20) | (df['Age'] > 40 )]
```

```python
# Q17. select * from cust order by Score desc limit 10

df.sort_values(by = 'Score', ascending=False)
```

```python
# Q18. select * from cust order by Score desc limit 10 offset 5

df.sort_values(by = 'Score', ascending=False)[5:15]
```
